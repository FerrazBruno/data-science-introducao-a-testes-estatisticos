# Analisando o conjunto de dados do MovieLens

```python
# importando os dados
notas = pd.read_csv('ratings.csv')
notas.head()
```

```python
# agrupando as notas por filme
nota_media_por_filme = notas.groupby('movieId').mean()['rating']
nota_media_por_filme.head()
```

```python
# visualisando as médias dos filmes
ax = sns.distplot(nota_media_por_filme.values)
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de votos por filmes do MovieLens')
```

```python
# contando a quantidade de votos por filme
quantidade_de_votos_por_filme = notas.groupby('movieId').count()
quantidade_de_votos_por_filme
```

```python
# selecionando os filmes com 10 ou mais votos
filmes_com_pelo_menos_10_votos = quantidade_de_votos_por_filme.query('userId >= 10').index
filmes_com_pelo_menos_10_votos
```

```python
# extraindo os valores de filmes_com_pelo_menos_10_votos
filmes_com_pelo_menos_10_votos.values
```

```python
# obtendo a nota média dos filmes com mais de 10 votos
nota_media_dos_filmes_com_pelo_menos_10_votos = nota_media_por_filme.loc[filmes_com_pelo_menos_10_votos.values]
nota_media_dos_filmes_com_pelo_menos_10_votos.head()
```

```python
# plotando o gráfico
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos)
ax.set(xlabel='Nota média', ylabel='Densidade')
ax.set_title('Média de votos em filmes no MovieLens')
```

```python
# visualizando a distribuição dos filmes através de um box-plot
ax = sns.boxplot(x=nota_media_dos_filmes_com_pelo_menos_10_votos)
ax.set(xlabel='Nota média dos filmes')
ax.set_title('Distribuição de nota média dos filmes avaliados no MovieLens')
```

```python
# visualizando os gráficos do MovieLens de forma acumulada para analisar se um filme é realmente bom ou não baseados na quantidade de avaliações 
ax = sns.distplot(nota_media_dos_filmes_com_pelo_menos_10_votos,
                  hist_kws={'cumulative':True}, kde_kws={'cumulative':True})
ax.set(xlabel='Nota média', ylabel='Proporção acumulada de filmes')
ax.set_title('Média de votos em filmes no MovieLens')
```

```python
# Analisando o comportamento da média de um filme
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

medias = [temp[0:i].mean() for i in range(1, len(temp))]

plt.plot(medias)
```

```python
# Calculando o ztest para cada uma das amostras
np.random.seed(75241)
temp = notas1.sample(frac=1).rating

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p = ztest(temp[0:i], value=nota_media_dos_filmes_com_pelo_menos_10_votos.mean())
    return (i, media, p)

medias = [calcula_teste(i) for i in range(2, len(temp))]
medias
```

```python
# Transformando as médias em um array do numpy
medias = np.array([calcula_teste(i) for i in range(2, len(temp))])
medias
```

```python
# Plotando as médias
plt.plot(medias[0:,1])
```

```python
# Renomeando a variável média
valores = np.array([calcula_teste(i) for i in range(2, len(temp))])
valores
```

```python
# Plotando valores 
plt.plot(valores[:,0], valores[:,1])
```

```python
# Plotando as variações do p-value
plt.plot(valores[:,0], valores[:,1])
plt.plot(valores[:,0], valores[:,2])
```

```python
# Delimitando uma linha referencial ao longo do eixo y de acordo com o valor 0.05
plt.figure(figsize=(12, 8))
plt.plot(valores[:,0], valores[:,1])
plt.plot(valores[:,0], valores[:,2])
plt.hlines(y=0.05, xmin=2, xmax=len(temp), colors='red')
```

> Temos duas distribuições bem mais parecidas, o que poderia nos levar a conclusões precipitadas sobre os nossos dados.

```python
# Aprofundando no cenário de poucos elementos
descr_todas_as_notas = DescrStatsW(notas.rating)
descr_toy_story = DescrStatsW(notas1[3:12].rating)
comparacao = descr_todas_as_notas.get_compare(descr_toy_story)

comparacao.summary()
```

> Como nosso p value é 0.877, não podemos descartar a hipótese nula (de que as médias desses conjuntos são iguais). Já o intervalo de confiança vai de -0.735 até 0.627 - um intervalo discrepante e que não permite sabermos se uma média é maior ou menor do que a outra, impedindo, também, que tomemos uma decisão em relação a essa comparação. Dessa forma, nossas únicas alternativas são assumirmos que elas são iguais ou que simplesmente não temos dados suficientes para essa análise.

# Comparar a média entre dois filmes

```python
# Selecionando os filmes
filmes.query('movieId in [1, 593, 72226]')
```

```python
# Separando as notas de cada filme
notas1 = notas.query('movieId == 1')
notas593 = notas.query('movieId == 593')
notas72226 = notas.query('movieId == 72226')
```

```python
# Plotando o boxplot dos filmes
plt.figure(figsize=(12, 8))
plt.boxplot([notas1.rating, notas593.rating, notas72226.rating], labels=['Toy Story', 'Silence of the Lambs', 'Fantastic Mr. Fox'])
plt.title('Distribuição das notas de acordo com o filmes')
```

> Nos boxplots gerados, já poderemos perceber, visualmente, que tanto Toy Story quanto Silence of the Lambs possuem uma mediana 4. Já o filme Fantastic Mr. Fox tem uma mediana um pouco mais elevada, mas que não conseguimos apreender a olho.

```python
# Verificando os valores do Fantastic Mr. Fox
notas72226.describe()    
```

```python
# Gerando o boxplot pelo seaborn
sns.boxplot(x='movieId', y='rating', data=notas.query('movieId in [1, 593, 72226]'))
```

```python
# Estipulando qual dos 3 filmes é o melhor
descr_1 = DescrStatsW(notas1.rating)
descr_593 = DescrStatsW(notas593.rating)
comparacao = descr_1.get_compare(descr_593)

comparacao.summary()
```

> Aplicando o Teste Z com alpha = 0.05, encontramos uma diferença estatística entre os filmes Toy Story e Silence of the Lambs, com um p value igual a 0.02 e um intervalo de confiança entre -0.391 e -0.090

```python
# Comparando Silent of the Lambs e Fantastic Mr. Fox
descr_72226 = DescrStatsW(notas72226.rating)
descr_593 = DescrStatsW(notas593.rating)
comparacao = descr_72226.get_compare(descr_593)

comparacao.summary()
```

> Dessa vez, nosso p value não indica uma diferença significante (pois não é menor do que 0.05), e o intervalo inclui tanto a chance do filme ser pior, quanto a chance dele ser melhor.

```python
# Comparando Toy Story e Fantastic Mr. Fox
comparacao = descr_1.get_compare(descr_72226)
comparacao.summary()
```

> Da mesma forma, não temos uma diferença significativa.

```python
# Utilizando o normaltest para descobrir se os dados derivam de uma normal
from scipy.stats import normaltest

_, p = normaltest(notas1.rating)
p
```

> Teremos um p value de 0.00011053430732728716 - ou seja, menor do que 0.05. Como essa função testa a hipótese nula de que o conjunto de dados venha de uma distribuição normal, um resultado menor que 0.05 indica que não estamos trabalhando com uma distribuição normal.

```python
# Comparando conjuntos de dados
from scipy.stats import ranksums

_, p = ranksums(notas1.rating, notas593.rating)
p
```

> Nesse caso, o p value é significativo, descartando a hipótese nula. No ranksums(), a hipótese nula é de que as duas amostras vieram da mesma distribuição. Já a hipótese alternativa, que admitiremos nesse caso (já que o p value foi bem pequeno), é a de que os valores de uma amostra têm uma tendência a serem maiores que os da outra amostra.
